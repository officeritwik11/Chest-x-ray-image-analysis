```python
from fastai.transforms import *
from fastai.torch_imports import *
from fastai.conv_learner import *
from fastai.model import *
from fastai.dataset import *
from fastai.sgdr import *
from fastai.plots import *
```

```python
#architecture, dataset Path, size of each image.
PATH = "data/lungXray/chest_xray/"
sz = 500
arch = resnet50
```

```python
torch.cuda.is_available()
```

```python
torch.backends.cudnn.enabled
```

```python
#Checking for folders in PATH variable
os.listdir(PATH)
```

```python
trainImages = os.listdir(f'{PATH}train/NORMAL/')
print(trainImages[:10])
print('\nNo. of training images =',len(trainImages))
```

```python
#Reading an image from Training folder of NORMAL subfolder.

imageSample = plt.imread(f'{PATH}train/NORMAL/{trainImages[20]}')
print(imageSample)
plt.imshow(imageSample)
```

```python
data = ImageClassifierData.from_paths(path=PATH,bs=28,tfms=tfms_from_model(arch,sz))
learn = ConvLearner.pretrained(arch,data,precompute=False)
learn.load('500_lungXray')
learn.fit(1e-3,3)
```

```python
# finding optimal LR using cyclic 
lrf = learn.lr_find()
```

```python
learn.sched.plot_lr()
```

```python
learn.sched.plot()
```

```python
'''
 From this above graph it can be said that we are getting LR as 1e-3 where there is sudden decrease 
 in LR just before diverging.
'''
learn.fit(1e-3,2)
```

```python
# clearly we can see our model is overfitting over here,
#Let's create augmented data by apply certain transformations.

tfms = tfms_from_model(arch,sz)
data = ImageClassifierData.from_paths(PATH,bs=64,tfms=tfms,num_workers=1)
learn = ConvLearner.pretrained(arch,data,precompute=True)
```

```python
learn.fit(1e-3,3)
```

```python
#applying cyclic gradient descent
data = ImageClassifierData.from_paths(PATH,bs=28,tfms=tfms,num_workers=1)
learn = ConvLearner.pretrained(arch,data,precompute=False)
```

```python
learn.fit(1e-3,3,cycle_len=1)
```

```python
learn.sched.plot_lr()
```

```python
learn.unfreeze()
```

```python
singleLR = 1e-3
lr = np.array([singleLR/10,singleLR/9,singleLR])
```

```python
data = ImageClassifierData.from_paths(PATH,bs=28,tfms=tfms,num_workers=1)
learn = ConvLearner.pretrained(arch,data,precompute=False)
learn.fit(lr,2,cycle_len=1,cycle_mult=2)
```

```python
#saving these weights:
learn.save('500_lungXray')
```

```python
learn.load('500_lungXray')
data = ImageClassifierData.from_paths(PATH,bs=28,tfms=tfms,num_workers=1)
learn = ConvLearner.pretrained(arch,data,precompute=True)
```

```python
learn.fit(1e-3,3)
```

```python
log_preds,y = learn.TTA()   #gives actual values.
probs = np.mean(np.exp(log_preds),0) #gives actual probability.
```

```python
accuracy_np(probs, y)
```

```python
preds = np.argmax(probs, axis=1)
probs=probs[:,1]
```

```python
probs.shape
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y, preds)
```

```python
plot_confusion_matrix(cm, data.classes)
```
